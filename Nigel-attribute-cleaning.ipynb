# Attribute cleaning

```python
PATH = 'Data/attribute-data/img/'
```

```python
cat_names = pd.read_csv('Data/attribute-data/list_category.txt',sep ='\s+',header=None).reset_index().rename(columns={0:'category',1:'category_type','index':'cat_num'})

```

```python
cat_names['cat_num'] = cat_names['cat_num'].apply(lambda x : x+1)
```

```python
cat_names.head()
```

```python
cat_num = pd.read_csv('Data/attribute-data/cat.txt',sep='\s+',header=None,names=['category']).reset_index().rename(columns={'category':'cat_num','index':'img'})

```

```python
cat_num[['img']] = cat_num[['img']].applymap(lambda x: x[4:])
```

```python
cat_num.head()
```

```python
cat = cat_num.merge(cat_names, on='cat_num', how='left').drop(columns='cat_num').set_index('img')

```

```python
cat
```

```python
cat['category_type'] = cat['category_type'].map({1:'upper',2:'lower',3:'full body'})
```

```python
attr_names = pd.read_csv('Data/attribute-data/list_attr_simple.txt',sep='\s+', header=None).rename(columns={0:'attribute',1:'attribute type'})
```

```python
attr_names
```

```python
attr_names['attribute type'] = attr_names['attribute type'].map({1:'design',2:'sleeves',3:'length',4:'part',5:'fabric',6:'fit'})
```

```python
attr_names
```

```python
attr_names_headers = attr_names.iloc[:,0]
attr_names_headers
```

```python
attr_tags = pd.read_csv('Data/attribute-data/tags_attr_simple.txt',sep='\s+',header=None,names=attr_names_headers)
```

```python
attr_tags
```

```python
attr_img = pd.read_csv('Data/attribute-data/img_attr_simple.txt',sep='\s+',header=None).rename(columns={0:'img'})

```

```python
attr_img
```

```python
attr = attr_img.join(attr_tags,how='inner')
```

```python
attr['img'] = attr['img'].apply(lambda x: x[4:])
```

```python
attr
```

```python
bb = pd.read_csv('Data/attribute-data/bbox.txt',sep='\s+',header=None,index_col=False, names=['img','x_1', 'y_1', 'x_2', 'y_2'])
```

```python
bb['img'] = bb['img'].apply(lambda x: x[4:])
```

```python
bb.set_index('img',inplace=True)
```

```python
bb
```

```python
pd.set_option('display.max_columns', None)
```

```python
landmarks = pd.read_csv('Data/attribute-data/landmarks.txt',sep='\s+',names=['img','clothes_type','v1','x1','y1','v2','x2','y2','v3','x3','y3','v4','x4','y4','v5','x5','y5','v6','x6','y6','v7','x7','y7','v8','x8','y8']).fillna(0).drop(columns=['clothes_type'])
```

```python
landmarks['img'] = landmarks['img'].apply(lambda x: x[4:])
```

```python
landmarks.set_index('img',inplace=True)
```

```python
cat = cat.reset_index()
```
